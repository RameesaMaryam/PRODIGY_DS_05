# **Analyze traffic accident data to identify patterns related to road conditions, weather, and time of day. Visualize accident hotspots and contributing factors.**

# **Dataset :-**https://www.kaggle.com/datasets/ggsri123/realtime-traffic-incident-reports



```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Austin_Traffic_Accidents.csv')

# Basic data exploration
print(df.head())  # Print the first few rows to understand the structure of the data
print(df.info())  # Get an overview of columns and data types

# Data preprocessing
# Convert 'Published Date' column to datetime format
df['Published Date'] = pd.to_datetime(df['Published Date'])

# Example 1: Plotting accident counts by hour of day
plt.figure(figsize=(10, 6))
sns.countplot(x=df['Published Date'].dt.hour, data=df, palette='viridis')
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Example 2: Visualizing accident hotspots using latitude and longitude
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Longitude', y='Latitude', data=df, alpha=0.6)
plt.title('Accident Hotspots in Austin')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.tight_layout()
plt.show()

# Example 3: Visualizing accident counts by issue reported
plt.figure(figsize=(10, 6))
sns.countplot(y='Issue Reported', data=df, order=df['Issue Reported'].value_counts().index[:10], palette='pastel')
plt.title('Accidents by Issue Reported in Austin')
plt.xlabel('Number of Accidents')
plt.ylabel('Issue Reported')
plt.tight_layout()
plt.show()
```
